```python
# https://matplotlib.org/stable/users/explain/quick_start.html#quick-start
# 그림보고 어떤걸 수정하려고 하는지 확인

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2 * np.pi, 200)
y = np.sin(x)

fig, ax = plt.subplots() # figure는 흰도화지, 밑에 ax(그래프)를 만듦

## fig, ax = plt.subplots()       # a figure with a single Axes
## fig, axs = plt.subplots(2, 2)  # a figure with a 2x2 grid of Axes -- 각각의 xlabel, ylabel에 이름을 주거나 할 수 있음

ax.plot(x, y)
plt.show()
```

```python
fig,ax = plt.subplots()

# 각각의 plot에서 alpha 매개변수를 사용하여 투명도 조절
# legend 달기
ax.plot(x, x+1, '-g', alpha=0.5, label="Green")  # 초록색 실선, 투명도 0.5
ax.plot(x, x+2, '--c', alpha=0.7, label="Blue")  # 청록색 대시선, 투명도 0.7
ax.plot(x, x+3, '-.k', lw=7, alpha=0.3, label="Black")  # 검정색 점선 점선, 두께 7, 투명도 0.3
ax.plot(x, x+4, ':r', lw=5, alpha=0.9, label="Red")  # 빨간색 점선, 두께 5, 투명도 0.9
ax.legend()
plt.show()
```

```python
# 3차원 함수 시각화하기
def f(x,y):
    return np.sin(x)**10 + np.cos(10+y*x)*np.cos(x)
```

```python
x = np.linspace(0,5,50)
y = np.linspace(0,5,40)

X, Y = np.meshgrid(x,y)
Z = f(X,Y)
plt.contour(X,Y,Z, colors="black")
```

```python
plt.contour(X,Y,Z, 20, cmap="RdGy")
plt.colorbar();
```

```python
def grayscale_cmap(cmap):
    # 색상 지도의 회색 조 버전을 반환
    cmap = plt.cm.get_cmap(cmap)
    colors = cmap(np.arange(cmap.N))

    # RGBA를 감지할 수 있는 회색 조 밝기로 전환
    RGB_weight = [0.299,0.587,0.114]
    luminance = np.sqrt(np.dot(colors[:,:3]**2,RGB_weight))
```

```python
from sklearn.datasets import load_digits
digits = load_digits(n_class=6)

fig,ax = plt.subplots(8,8,figsize=(6,6))
for i, axi in enumerate(ax.flat):
    axi.imshow(digits.images[i],cmap='binary')
    axi.set(xticks=[],yticks=[])
```

```python
from sklearn.manifold import Isomap
iso = Isomap(n_components=2)
projection = iso.fit_transform(digits.data)
```

```python
# 산점도 그리기
plt.scatter(projection[:, 0], projection[:, 1], lw=0.1,
            c=digits.target, cmap=plt.cm.get_cmap('cubehelix', 6))
plt.colorbar(ticks=range(6), label='digit value')
plt.clim(-0.5, 5)  # plt.clim()에서 3개의 인수를 넘길 필요가 없습니다.

plt.show()
```
